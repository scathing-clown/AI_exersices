# Задание 3.1

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
import scipy.stats as stats

df = pd.read_csv('https://raw.githubusercontent.com/akmand/datasets/master/iris.csv')
df.head()
```

```python
feature1 = [[x] for x in df.sepal_length_cm]
feature1 = np.array(feature1)
minmax_scale = preprocessing.MinMaxScaler(feature_range = (0,1))
scaled_feature1 = minmax_scale.fit_transform(feature1)
df1 = df
df1.sepal_length_cm = scaled_feature1
df1.head()
```

```python
feature2 = [[x] for x in df.sepal_width_cm]
feature2 = np.array(feature2)
scaled_feature2 = stats.zscore(feature2)
df2 = df1
df2.sepal_width_cm = scaled_feature2
df2.head()
```
