# Задание 4.3

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
import math
from scipy.integrate import simps
from numpy import trapz
%matplotlib inline

aq = []
for x in range(10):
    y = abs(math.cos(x*np.exp(math.cos(x)+math.log(x+1))))
    aq.append(y)
x = np.arange(0,2*np.pi,0.2*np.pi)
plt.plot(x, aq, c = 'r')
plt.fill_between(x, aq)
plt.show()

area = trapz(aq)
print(area)
```

```python
while True:
    print ("Операция или функция?")
    answ = str(input())
    if answ == 'Операция':
        s = input("Знак (+,-,*,/): ")
        if s in ('+', '-', '*', '/'):
            a = float(input("a="))
            b = float(input("b="))
            if s == '+':
                print("%.2f" % (a+b))
            if s == '-':
                print("%.2f" % (a-b))
            if s == '*':
                print("%.2f" % (a*b))
            if s == '/':
              if b != 0:
                  print("%.2f" % (a/b))
              else:
                  print("Нельзя делить на ноль!")

        else:
          print("Неверно")
    elif answ == 'Функция':
        i = input ("Функция e^(a+b), sin(a+b), cos(a+b), a^b (1, 2, 3, 4)")
        a = float(input("a="))
        b = float(input("b="))
        if i==1:
            print(math.e**(a+b))
        elif i == 2:
            print(math.sin(a+b))
        elif i == 3:
            print(math.cos(a+b))
        else:
            print(a**b)
    elif answ == 'Стоп':
        break
    else:
        print("Неверная команда")
```
